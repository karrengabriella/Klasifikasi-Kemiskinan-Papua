```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur pandas untuk menampilkan semua baris dan kolom
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Data kemiskinan
data_kemiskinan = {
    'Kabupaten': ['Provinsi Papua', 'Merauke', 'Jayawijaya', 'Jayapura', 'Nabire',
                  'Kepulauan Yapen', 'Biak Numfor', 'Paniai', 'Puncak Jaya', 'Mimika',
                  'Boven Digoel', 'Mappi', 'Asmat', 'Yahukimo', 'Pegunungan Bintang',
                  'Tolikara', 'Sarmi', 'Keerom', 'Waropen', 'Supiori', 'Mamberamo Raya',
                  'Nduga', 'Lanny Jaya', 'Mamberamo Tengah', 'Yalimo', 'Puncak', 'Dogiyai',
                  'Intan Jaya', 'Deiyai', 'Kota Jayapura'],
    '2021': [26.86, 10.16, 37.09, 12.13, 23.83, 26.09, 24.45, 36.59, 36.00, 14.17,
             19.90, 26.05, 24.83, 37.64, 30.46, 32.60, 13.84, 16.00, 29.85, 37.91,
             28.78, 37.18, 38.73, 36.76, 33.25, 36.26, 28.81, 41.66, 40.59, 11.39],
    '2022': [26.56, 10.10, 35.80, 11.94, 23.90, 26.15, 24.48, 36.63, 36.30, 14.28,
             20.04, 26.04, 24.87, 36.21, 30.32, 31.31, 13.76, 15.77, 30.22, 37.70,
             30.73, 36.73, 37.76, 35.92, 31.68, 36.25, 29.48, 42.03, 40.31, 11.12],
    '2023': [26.03, 10.01, 34.71, 11.45, 23.35, 25.89, 23.53, 35.39, 35.60, 13.55,
             19.80, 25.72, 24.36, 36.08, 29.79, 31.57, 13.21, 15.68, 29.16, 36.99,
             29.63, 37.09, 36.94, 35.42, 30.97, 36.44, 29.20, 40.01, 38.66, 10.50]
}

# Data pengangguran
data_pengangguran = {
    'Kabupaten': ['Provinsi Papua', 'Merauke', 'Jayawijaya', 'Jayapura', 'Nabire', 'Kepulauan Yapen',
                  'Biak Numfor', 'Paniai', 'Puncak Jaya', 'Mimika', 'Boven Digoel',
                  'Mappi', 'Asmat', 'Yahukimo', 'Pegunungan Bintang', 'Tolikara',
                  'Sarmi', 'Keerom', 'Waropen', 'Supiori', 'Mamberamo Raya',
                  'Nduga', 'Lanny Jaya', 'Mamberamo Tengah', 'Yalimo', 'Puncak',
                  'Dogiyai', 'Intan Jaya', 'Deiyai', 'Kota Jayapura'],
    '2021': [3.33, 2.19, 1.62, 7.87, 3.05, 2.47, 9.49, 0.74, 1.74, 5.37, 6.73, 4.11, 0.45,
             2.25, 4.43, 1.12, 3.5, 1.41, 8.56, 2.66, 1.91, 0, 0, 0.9, 0.41, 0.94,
             5.68, 1.43, 0.79, 11.67],
    '2022': [2.83, 2.64, 0.45, 5.94, 3.45, 3.6, 7.4, 1.83, 0.09, 5.49, 5.56, 2.34, 0.87,
             1.32, 3.01, 0.36, 2.09, 2.49, 6.64, 2.93, 2.32, 0.52, 0.56, 0.95, 0.73,
             0.19, 4.54, 2.14, 1.48, 9.71],
    '2023': [2.67, 3.61, 0.51, 4.07, 4.06, 4.01, 5.72, 1.25, 0.73, 4, 3.07, 4.22, 2.14,
             2.54, 1.58, 0.73, 3.09, 1.5, 4.62, 4.81, 0.4, 0.05, 0.3, 0.96, 0.24,
             0.24, 2.39, 3.22, 1.48, 10.76]
}

# Convert data to DataFrames
df_kemiskinan = pd.DataFrame(data_kemiskinan)
df_pengangguran = pd.DataFrame(data_pengangguran)

# Menggabungkan data kemiskinan dan pengangguran berdasarkan kabupaten dan tahun
df_kemiskinan_melted = df_kemiskinan.melt(id_vars='Kabupaten', var_name='Tahun', value_name='Kemiskinan')
df_pengangguran_melted = df_pengangguran.melt(id_vars='Kabupaten', var_name='Tahun', value_name='Pengangguran')

# Menggabungkan DataFrame
df_combined = pd.merge(df_kemiskinan_melted, df_pengangguran_melted, on=['Kabupaten', 'Tahun'])

# Skala data sebelum melakukan clustering
scaler = StandardScaler()
data_clustering_scaled = scaler.fit_transform(df_combined[['Kemiskinan', 'Pengangguran']])

# Menampilkan hasil normalisasi per tahun
df_combined_scaled = pd.DataFrame(data_clustering_scaled, columns=['Kemiskinan', 'Pengangguran'])
df_combined_scaled['Kabupaten'] = df_combined['Kabupaten']
df_combined_scaled['Tahun'] = df_combined['Tahun']

print("Hasil Normalisasi per Tahun:")
print(df_combined_scaled[['Kabupaten', 'Tahun', 'Kemiskinan', 'Pengangguran']].to_string(index=False))

# Tentukan jumlah klaster optimal dengan Yellowbrick Elbow Visualizer
model = KMeans(random_state=42, max_iter=100)
visualizer = KElbowVisualizer(model, k=(1, 10), timings=False)

visualizer.fit(data_clustering_scaled)    # Fit the data to the visualizer
visualizer.show()                         # Finalize and render the figure

# Jumlah klaster optimal
optimal_clusters = visualizer.elbow_value_

# Lakukan klasterisasi per tahun dengan jumlah klaster optimal
kmeans_per_year = KMeans(n_clusters=optimal_clusters, random_state=42, max_iter=100)
kmeans_per_year.fit(data_clustering_scaled)
df_combined['Cluster'] = kmeans_per_year.labels_

# Menampilkan hasil clustering per tahun
print("Hasil Clustering per Tahun:")
print(df_combined[['Kabupaten', 'Tahun', 'Kemiskinan', 'Pengangguran', 'Cluster']].to_string(index=False))

# Menampilkan centroid per iterasi
centroids = kmeans_per_year.cluster_centers_
print("\nCentroid Klaster per Tahun:")
for i, centroid in enumerate(centroids):
    print(f"Centroid {i+1}: {centroid}")

print(f"\nJumlah iterasi yang terjadi dalam klasterisasi per tahun: {kmeans_per_year.n_iter_}")

# Visualisasi hasil
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_combined, x='Kemiskinan', y='Pengangguran', hue='Cluster', palette='viridis')
plt.title('Klasterisasi Kemiskinan dan Pengangguran')
plt.xlabel('Presentase Kemiskinan')
plt.ylabel('Presentase Pengangguran')
plt.show()

# Klasterisasi final berdasarkan rata-rata per daerah
df_agg = df_combined.groupby('Kabupaten').agg({
    'Kemiskinan': 'mean',
    'Pengangguran': 'mean'
}).reset_index()

# Skala data sebelum melakukan clustering final
data_clustering_final_scaled = scaler.fit_transform(df_agg[['Kemiskinan', 'Pengangguran']])

# Menampilkan hasil normalisasi per daerah
df_agg_scaled = pd.DataFrame(data_clustering_final_scaled, columns=['Kemiskinan', 'Pengangguran'])
df_agg_scaled['Kabupaten'] = df_agg['Kabupaten']

print("Hasil Normalisasi per Daerah:")
print(df_agg_scaled[['Kabupaten', 'Kemiskinan', 'Pengangguran']].to_string(index=False))

# Tentukan jumlah klaster optimal untuk klasterisasi final dengan Yellowbrick Elbow Visualizer
visualizer_final = KElbowVisualizer(model, k=(1, 10), timings=False)

visualizer_final.fit(data_clustering_final_scaled)    # Fit the data to the visualizer
visualizer_final.show()                               # Finalize and render the figure

# Jumlah klaster optimal untuk klasterisasi final
optimal_clusters_final = visualizer_final.elbow_value_

# Lakukan klasterisasi final dengan jumlah klaster optimal
kmeans_final = KMeans(n_clusters=optimal_clusters_final, random_state=42, max_iter=100)
kmeans_final.fit(data_clustering_final_scaled)
df_agg['Cluster'] = kmeans_final.labels_

# Menampilkan hasil clustering final
print("Hasil Clustering Final:")
print(df_agg[['Kabupaten', 'Kemiskinan', 'Pengangguran', 'Cluster']].to_string(index=False))

# Menampilkan centroid klaster final
centroids_final = kmeans_final.cluster_centers_
print("\nCentroid Klaster Final:")
for i, centroid in enumerate(centroids_final):
    print(f"Centroid {i+1}: {centroid}")

print(f"\nJumlah iterasi yang terjadi dalam klasterisasi final: {kmeans_final.n_iter_}")

# Visualisasi hasil klasterisasi final
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_agg, x='Kemiskinan', y='Pengangguran', hue='Cluster', palette='viridis', legend='full')
plt.title('Klasterisasi Final Kemiskinan dan Pengangguran')
plt.xlabel('Presentase Kemiskinan')
plt.ylabel('Presentase Pengangguran')
plt.legend(title='Klaster')
plt.show()
```
